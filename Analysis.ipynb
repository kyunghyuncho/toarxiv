```python
%matplotlib notebook

import numpy
import matplotlib.pyplot as plot
import matplotlib

matplotlib.rcParams.update({'font.size': 22})
matplotlib.rcParams.update({'font.family': 'serif'})
```

```python
data = numpy.loadtxt(open("./survey.csv","rb"),delimiter=",",skiprows=1,dtype=numpy.object)
```

```python
answers = [numpy.unique(data[:, ii]) for ii in xrange(data.shape[1])]
```

```python
portions = [len(numpy.where(data[:, 0] == aa)[0]) for aa in answers[0]]
portions = numpy.array(portions, dtype='float')/numpy.sum(portions)
print portions, answers[0]

f = plot.figure()
ax = f.gca()
ax.pie(portions, labels=['%2.1f%%'%(p*100) for p in portions])
plot.legend(['Not to arXiv', 'To arXiv'], loc='lower right')
plot.title('')
plot.show()
```

```python
portions = [len(numpy.where(data[:, 2] == aa)[0]) for aa in answers[2]]
portions = numpy.array(portions, dtype='float')/numpy.sum(portions)
print portions, answers[2]

f = plot.figure()
ax = f.gca()
ax.pie(portions, labels=['%2.1f%%'%(p*100) for p in portions])
plot.legend(answers[2], loc='best', fontsize=18, bbox_to_anchor=(0.7,0.1))
plot.title('')
plot.show()
```

```python
portions = [len(numpy.where(data[numpy.where(data[:,2]=='Machine Learning')[0], 0] == aa)[0]) 
            for aa in answers[0]]
portions = numpy.array(portions, dtype='float')/numpy.sum(portions)
print portions, answers[0]

f = plot.figure()
ax = f.gca()
ax.pie(portions, labels=['%2.1f%%'%(p*100) for p in portions])
plot.legend(['Not to arXiv', 'To arXiv'], loc='lower right')
plot.title('')
plot.show()
```
